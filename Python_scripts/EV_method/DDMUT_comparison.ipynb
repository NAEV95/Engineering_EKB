```python
%pip install scikit-learn

from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error
import numpy as np

# Assuming delta_30 and ddmut_prediction are columns in the dataframe df
delta_30 = df['delta_30']
ddmut_prediction = df['ddmut_prediction']



r2 = r2_score(delta_30, ddmut_prediction)
rmse = mean_squared_error(delta_30, ddmut_prediction, squared=False)
mae = mean_absolute_error(delta_30, ddmut_prediction)
absolute_error = np.abs(delta_30 - ddmut_prediction)
absolute_error_std = np.std(absolute_error)
standard_error_std = absolute_error_std / np.sqrt(len(delta_30))

r2, rmse, mae, absolute_error_std, standard_error_std
```
